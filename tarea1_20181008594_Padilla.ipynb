## 1 items del para llenar el DataFrame


```python
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Itemsproductos = ['Iphone 13 Pro Max', 'Iphone 14 Pro Max', 'Iphone 15 Pro Max', 'Iphone 11 Pro','Iphone X Pro']
Itemsvendedores = ['Juan Jose', 'Belinda Sofia', 'David Alejandro', 'Pedro Leonel']
Itemsregiones = ['Centro', 'Anillo Periferico', 'BFFA', 'La Laguna']
```

## 2 Generado de Datos Aleatorios

```python
cantidad_datos=3000
```

```python
producto = random.choices(productos, k=cantidad_datos)
vendedor= random.choices(vendedores, k=cantidad_datos)
region= random.choices(regiones, k=cantidad_datos)

cantidad=[]
precio_unitario=[]
precio_total=[]

for i in range(cantidad_datos):
    cantidad.append( random.randint(1, 100))

for i in range(cantidad_datos):
    precio_unitario.append( round(random.uniform(5.0, 100.0), 2))
    
fecha= pd.date_range(start='2023-01-01', periods=cantidad_datos, freq='D').tolist()

for i in range(cantidad_datos):
    precio_total.append(cantidad[i]* precio_unitario[i])
```

## 3 Creacion del DataFrame

```python
df = pd.DataFrame({
    'producto': producto,
    'vendedor': vendedor,
    'region': region,
    'cantidad': cantidad,
    'precio_unitario': precio_unitario,
    'fecha': fecha,
    'precio_total': precio_total
})
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(df['cantidad'], bins=30, kde=True)
plt.title('DistribuciÃ³n de Cantidad')

plt.subplot(2, 2, 2)
sns.histplot(df['precio_unitario'], bins=30, kde=True)
plt.title('DistribuciÃ³n de Precio Unitario')

plt.subplot(2, 2, 3)
sns.histplot(df['precio_total'], bins=10, kde=True)
plt.title('DistribuciÃ³n de Precio Total')

plt.subplot(2, 2, 4)
sns.histplot(df['vendedor'] )
plt.title('DistribuciÃ³n de los Vendedores')

plt.tight_layout()
plt.show()
```
